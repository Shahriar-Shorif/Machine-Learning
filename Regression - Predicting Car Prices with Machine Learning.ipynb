```python
import pandas as pd
import numpy as np
url = 'https://raw.githubusercontent.com/rashakil-ds/Public-Datasets/main/automobile.csv'
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.replace("?", np.nan, inplace=True)
```

```python
df.columns
```

```python
numeric_columns = [
    'symboling', 'normalized-losses', 'wheel-base', 'length', 'width', 'height',
    'curb-weight', 'engine-size', 'bore', 'stroke', 'compression-ratio',
    'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price'
]
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce').astype(float)

    
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['price'], inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
df[['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm']] = imputer.fit_transform(
    df[['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm']]
)
```

```python
categorical_cols = ['make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style',
                    'drive-wheels', 'engine-location', 'engine-type',
                    'num-of-cylinders', 'fuel-system']
```

```python
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
X = df.drop(['price'], axis=1)
y = df['price']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model_lr = LinearRegression()
model_lr.fit(X_train, y_train)
```

```python
y_train_predict_lr = model_lr.predict(X_train)
y_test_predict_lr = model_lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
r2train = r2_score(y_train,y_train_predict_lr)
r2train
```

```python
r2test = r2_score(y_test,y_test_predict_lr)
r2test
```

```python
from sklearn.neighbors import KNeighborsRegressor
model_knn = KNeighborsRegressor(n_neighbors=5)
model_knn.fit(X_train, y_train)
```

```python
y_train_predict_knn = model_knn.predict(X_train)
y_test_predict_knn = model_knn.predict(X_test)
```

```python
r2knn = r2_score(y_train,y_train_predict_knn)
r2knn
```

```python
r2testknn = r2_score(y_test,y_test_predict_knn)
r2testknn
```
